```python
from pytket import Circuit
from pytket.qasm import circuit_to_qasm_str
```

```python
c = Circuit(1)
c.V(0)
c.measure_all()
```

```python
from pytket.utils import Graph
Graph(c).get_DAG()
```

```python
circuit_to_qasm_str(c)
```

```python
from qiskit.aqua.algorithms import Grover
from qiskit.aqua.components.oracles import TruthTableOracle
```

```python
grover = Grover(TruthTableOracle("0100"))
c = grover.construct_circuit(measurement=True)
```

```python
c.draw()
```

```python
from pytket.extensions.qiskit import IBMQBackend, qiskit_to_tk
backend = IBMQBackend("ibmq_qasm_simulator")
```

```python
c = qiskit_to_tk(c)
Graph(c).get_DAG()
```

```python
backend.compile_circuit(c)
Graph(c).get_DAG()
```

```python
circuit_to_qasm_str(c)
```
